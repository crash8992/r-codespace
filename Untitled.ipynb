```r
# Prueba básica de R en Jupyter
print("Hola, JupyterLab con R funciona 🚀")

# Crear un vector
x <- c(10, 20, 30, 40, 50)

# Calcular media y desviación estándar
cat("Media:", mean(x), "\n")
cat("Desviación estándar:", sd(x), "\n")

# Dibujar un gráfico simple
plot(x, type="b", col="blue", main="Gráfico de prueba en R", xlab="Índice", ylab="Valor")
```
